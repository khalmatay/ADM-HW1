```python
#Find a string
def count_substring(string, sub_string):
    s = ""
    count = 0
    for el in string:
        s += el
        if sub_string in s:
            count += 1
        if len(s) == len(sub_string):
            s = s[1:]

    return count


if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()

    count = count_substring(string, sub_string)
    print(count)
```

```python
#Text Wrap

def wrap(string, max_width):
    arr = []
    i = 0
    while i < len(string):
        arr.append(string[i:i + max_width])
        i += max_width
    return "\n".join(arr)
```

```python
#The Minion Game
def minion_game(string):
    vowels = "AEIOU"
    kevin = 0
    stuart = 0
    length = len(string)

    for i in range(length):
        if string[i] in vowels:
            kevin += length - i
        else:
            stuart += length - i

    if kevin > stuart:
        print("Kevin", kevin)
    elif kevin < stuart:
        print("Stuart", stuart)
    else:
        print("Draw")


if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
#Merge the Tools!

def merge_the_tools(string, k):
    arr = []

    for i in range(0, len(string), k):
        arr.append(string[i:i + k])

    for el in arr:
        s = []
        for j in el:
            if j not in s:
                s.append(j)
        print("".join(s))

    # your code goes here


if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```

```python
#sWAP cASE

def swap_case(s):
    string = ""
    for i in range(len(s)):
        if s[i].isupper():
            string += s[i].lower()
        else:
            string += s[i].upper()

    return string

```
